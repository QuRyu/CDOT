### Data Pre-processing

```python
import pandas as pd 

data = pd.read_excel('./PerlCOdownload.xls')
data.head(5)
```

```python
# drop EST rows
data = data[data['Vendor ID'] != '-EST-']


# For auctions with n >= 2 bidders, if the number of rows for n-bidder auctions is not larger than 32,
# then remove all such data
N = 32

participants_numbers = pd.unique(data['ParticipantsCount'])
to_keep = [] 
for i in participants_numbers:
    if len(data[data['ParticipantsCount'] == i]) >= N:
        to_keep.append(i)

        
# remove auctions with one bidder 
if 1 in to_keep:
    to_keep.remove(1)
        
data = data[data['ParticipantsCount'].isin(to_keep)]


# rename and drop some columns 
data.rename(columns={"% of Engineer's Estimate": 'PercentOfEstimates'}, inplace=True)
data.drop(columns=['Vendor Name', 'Total Bid', '% of Low Bid'], inplace=True)


data.head(5)
```

```python
# clean data in PerecentOfEstimates column
idx = data['PercentOfEstimates'] > 8
data.loc[idx, 'PercentOfEstimates'] = data[idx].PercentOfEstimates / 100
```

```python
# save the data
data.to_csv('DataProcessed.csv')
```
